```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def get_vif(value):
    variance=[variance_inflation_factor(value,i) for i in range(x.shape[1])]
    vif=pd.DataFrame({'columns':x.columns,'VIF':variance})
    return vif
```

```python
vif=get_vif(x)
```

```python
vif.head()
```

> If VIF is more than 5 or 10 then, it is not good and those features are affecting model performance by inflating variance in regression coefficiants

```python
vif_greater_than_5=vif[vif.VIF>5] # this columnas are more correlated with other independent variables.
```
